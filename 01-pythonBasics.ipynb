#### Question 5: Demonstrate mutable vs immutable - List and Tuple

```python
# Create a list (mutable)
my_list = ["priyesh", 11, True, 3.14]
print("Original list:", my_list)

# Modify list element
my_list[0] = "postyesh"
print("Modified list:", my_list)

# Create a tuple (immutable)
my_tuple = ("priyesh", 12, False, 3.17)
print("\nOriginal tuple:", my_tuple)

# Attempting to modify tuple will give an error
# my_tuple[0] = 'postyesh'  # This will raise TypeError: 'tuple' object does not support item assignment
print("Tuple remains unchanged (immutable)")
```

#### Question 6: Demonstrate behavior of mutable and immutable arguments

```python
def mylist(lst):
    """Modify mutable argument (list)"""
    lst[1] = 9
    print("Inside function - List:", lst)

def mytuple(tup):
    """Immutable argument (tuple) cannot be modified"""
    # tup[1] = 13  # This would give an error
    print("Inside function - Tuple:", tup)

my_list = ["priyesh", 11, True, 3.14]
my_tuple = ("priyesh", 12, False, 3.17)

print("Before function calls:")
print("List:", my_list)
print("Tuple:", my_tuple)

mylist(my_list)
mytuple(my_tuple)

print("\nAfter function calls:")
print("List (changed):", my_list)  # List is modified
print("Tuple (unchanged):", my_tuple)  # Tuple remains same
```

#### Question 7: Perform basic arithmetic operations

```python
def add(x, y):
    print(f"The addition of {x} + {y} is {x+y}")

def subtr(x, y):
    print(f"The subtraction of {x} - {y} is {x-y}")

def multi(x, y):
    print(f"The multiplication of {x} * {y} is {x*y}")

def div(x, y):
    print(f"The division of {x} / {y} is {x/y}")

def mod(x, y):
    print(f"The modulo of {x} % {y} is {x%y}")

def pwr(x, y):
    print(f"The power of {x} ** {y} is {x**y}")

# Example with hardcoded values
num1 = 10
num2 = 3

add(num1, num2)
subtr(num1, num2)
multi(num1, num2)
div(num1, num2)
mod(num1, num2)
pwr(num1, num2)
```

#### Question 13: Find sum of all even numbers between 1 and 50

```python
sum_even = 0
for x in range(1, 51):
    if x % 2 == 0:
        sum_even += x

print(f"The sum of all even numbers between 1 and 50 is {sum_even}")
```

#### Question 14: Reverse a string using while loop

```python
string = "priyesh"  # Example string
i = len(string) - 1
rev = ""

while i >= 0:
    rev += string[i]
    i -= 1

print(f"Original string: {string}")
print(f"Reversed string: {rev}")
```

#### Question 15: Calculate factorial using while loop

```python
number = 5  # Example: change this value

if number == 0:
    print("The factorial of 0 is 1")
elif number < 0:
    print("The factorial of negative number is not defined")
else:
    factorial = 1
    i = 1
    while i <= number:
        factorial = factorial * i
        i = i + 1
    print(f"The factorial of {number} is {factorial}")
```
