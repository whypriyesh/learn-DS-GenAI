# Python Data Structures Exercises

This notebook contains 30 practical exercises on Python data structures including strings, lists, tuples, sets, and dictionaries.

#### Question 1: Create a string with your name and print it

```python
name = "priyesh"
print(name)
```

#### Question 2: Find the length of the string "Hello World"

```python
text = "Hello World"
print(f'The length of "{text}" is {len(text)}')
```

#### Question 3: Slice the first 3 characters from the string "Python Programming"

```python
string = "Python Programming"
new_string = string[0:3]
print(f"Original string: {string}")
print(f"First 3 characters: {new_string}")
```

#### Question 4: Convert the string "hello" to uppercase

```python
str2 = "hello"
print(f"Original: {str2}")
print(f"Uppercase: {str2.upper()}")
```

#### Question 5: Replace the word "apple" with "orange" in the string "I like apple"

```python
str3 = "I like apple"
str4 = str3.replace("apple", "orange")
print(f"Original: {str3}")
print(f"After replace: {str4}")
```

#### Question 6: Create a list with numbers 1 to 5 and print it

```python
my_list = []
for x in range(1, 6):
    my_list.append(x)
print(f"List: {my_list}")
```

#### Question 7: Append the number 10 to the list [1, 2, 3, 4]

```python
list2 = [1, 2, 3, 4]
print(f"Original list: {list2}")
list2.append(10)
print(f"After append(10): {list2}")
```

#### Question 8: Remove the number 3 from the list [1, 2, 3, 4, 5]

```python
list3 = [1, 2, 3, 4, 5]
print(f"Original list: {list3}")
list3.pop(2)  # Remove element at index 2 (which is 3)
print(f"After pop(2): {list3}")
```

#### Question 9: Access the second element in the list ['a', 'b', 'c', 'd']

```python
list4 = ["a", "b", "c", "d"]
print(f"List: {list4}")
print(f"Second element (index 1): {list4[1]}")
```

#### Question 10: Reverse the list [10, 20, 30, 40, 50]

```python
list5 = [10, 20, 30, 40, 50]
print(f"Original list: {list5}")
list5.reverse()
print(f"Reversed list: {list5}")
```

#### Question 11: Create a tuple with elements 100, 200, 300 and print it

```python
tup = (100, 200, 300)
print(f"Tuple: {tup}")
```

#### Question 12: Access the second-to-last element of tuple ('red', 'green', 'blue', 'yellow')

```python
tup2 = ("red", "green", "blue", "yellow")
print(f"Tuple: {tup2}")
print(f"Second-to-last element: {tup2[-2]}")
print(f"Also using [1:]: {tup2[1:]}")
```

#### Question 13: Find the minimum number in tuple (10, 20, 5, 15)

```python
tup3 = (10, 20, 5, 15)
print(f"Tuple: {tup3}")
print(f"Minimum value: {min(tup3)}")
```

#### Question 14: Find the index of element "cat" in tuple ('dog', 'cat', 'rabbit')

```python
tup4 = ("dog", "cat", "rabbit")
print(f"Tuple: {tup4}")
print(f"Index of 'cat': {tup4.index('cat')}")
```

#### Question 15: Create tuple with fruits and check if "kiwi" is in it

```python
tup5 = ("apple", "banana", "kiwi")
print(f"Tuple: {tup5}")
if "kiwi" in tup5:
    print(f"Yes, the tuple {tup5} has 'kiwi' in it")
else:
    print("Kiwi not found in the tuple")
```

#### Question 16: Create a set with elements 'a', 'b', 'c' and print it

```python
my_set = {"a", "b", "c"}
print(f"Set: {my_set}")
```

#### Question 17: Clear all elements from set {1, 2, 3, 4, 5}

```python
set2 = {1, 2, 3, 4, 5}
print(f"Original set: {set2}")
set2.clear()
print(f"After clear(): {set2}")
```

#### Question 18: Remove element 4 from set {1, 2, 3, 4}

```python
set3 = {1, 2, 3, 4}
print(f"Original set: {set3}")
set3.remove(4)
print(f"After remove(4): {set3}")
```

#### Question 19: Find union of sets {1, 2, 3} and {3, 4, 5}

```python
set4 = {1, 2, 3}
set5 = {3, 4, 5}
union_set = set4.union(set5)
print(f"Set 1: {set4}")
print(f"Set 2: {set5}")
print(f"Union: {union_set}")
```

#### Question 20: Find intersection of sets {1, 2, 3} and {2, 3, 4}

```python
set6 = {1, 2, 3}
set7 = {2, 3, 4}
intersection_set = set6.intersection(set7)
print(f"Set 1: {set6}")
print(f"Set 2: {set7}")
print(f"Intersection: {intersection_set}")
```

#### Question 21: Create dictionary with keys "name", "age", "city" and print it

```python
dic = {"name": "priyesh", "age": 11, "city": "indore"}
print(f"Dictionary: {dic}")
```

#### Question 22: Add a new key-value pair to dictionary

```python
dic2 = {"name": "John", "age": 25}
print(f"Original dictionary: {dic2}")
dic2["country"] = "USA"
print(f"After adding 'country': {dic2}")
```

#### Question 23: Access value for key "name" in dictionary

```python
dic3 = {"name": "Alice", "age": 30}
print(f"Dictionary: {dic3}")
print(f"Value for 'name': {dic3['name']}")
```

#### Question 24: Remove the key "age" from dictionary

```python
dic4 = {"name": "Bob", "age": 22, "city": "New York"}
print(f"Original dictionary: {dic4}")
dic4.pop("age")
print(f"After removing 'age': {dic4}")
```

#### Question 25: Check if key "city" exists in dictionary

```python
dic5 = {"name": "Alice", "city": "Paris"}
print(f"Dictionary: {dic5}")
if "city" in dic5:
    print(f"Yes, 'city' exists in {dic5}")
else:
    print("City is not present")
```

#### Question 29: Combine two dictionaries into one

```python
dic7 = {"name": "priyesh", "age": 11}
dic8 = {"city": "indore", "fav-lang": "python"}

print(f"Dictionary 1: {dic7}")
print(f"Dictionary 2: {dic8}")

dic7.update(dic8)
print(f"Combined dictionary: {dic7}")
```

#### Question 30: Convert list of strings into a set
